```python
import pandas as pd  # Import pandas library for data manipulation
import plotly.express as px  # Import plotly.express for creating interactive visualizations

# Load the dataset from the specified file path
file_path = '/content/drive/MyDrive/Assignment/Dataset salary 2024.csv'  # Path to the CSV file
data = pd.read_csv(file_path)  # Read the CSV file into a DataFrame

# Display the first few rows of the dataset to understand its structure
print(data.head())

# Create a bubble chart to visualize salary versus experience level
# 'experience_level' is plotted on the x-axis
# 'salary_in_usd' is plotted on the y-axis
# Bubble size is determined by 'salary_in_usd'
# Bubble color represents different 'job_title'
# Hovering over a bubble will show the 'job_title'
# Title and labels are set for clarity
bubble_chart = px.scatter(data,
                          x='experience_level',  # X-axis: Experience Level
                          y='salary_in_usd',  # Y-axis: Salary in USD
                          size='salary_in_usd',  # Bubble size: Salary in USD
                          color='job_title',  # Bubble color: Job Title
                          hover_name='job_title',  # Hover text: Job Title
                          title='Bubble Chart: Salary vs Experience Level',  # Chart title
                          labels={  # Labels for axes and legend
                              'experience_level': 'Experience Level',
                              'salary_in_usd': 'Salary in USD',
                              'job_title': 'Job Title'
                          })

# Display the bubble chart
bubble_chart.show()
```
