```python
import pandas as pd
import plotly.express as px

# Load the dataset
file_path = '/content/drive/MyDrive/Assignment/Dataset salary 2024.csv'  # Change this to the correct path
data = pd.read_csv(file_path)  # Read the CSV file into a DataFrame

# Prepare the data for Sunburst chart
# Group the data by 'company_location', 'job_title', and 'experience_level'
# Sum the 'salary_in_usd' for each group and reset the index
sunburst_data = data.groupby(['company_location', 'job_title', 'experience_level'])['salary_in_usd'].sum().reset_index()

# Create the Sunburst chart using Plotly Express
# Specify the path to define the hierarchy and the values to determine the size of each segment
sunburst_chart = px.sunburst(
    sunburst_data,  # Data for the sunburst chart
    path=['company_location', 'job_title', 'experience_level'],  # Define the hierarchy of the sunburst chart
    values='salary_in_usd',  # Define the values for the size of each segment
    title='Sunburst Chart: Distribution of Salaries by Location, Job Title, and Experience Level',  # Title of the chart
    color='salary_in_usd',  # Define the color of each segment based on salary
    color_continuous_scale='RdBu'  # Color scale to use for the chart
)

# Display the sunburst chart
sunburst_chart.show()
```
