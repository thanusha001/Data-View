```python
import pandas as pd  # Import pandas library for data manipulation
import seaborn as sns  # Import seaborn library for statistical data visualization
import matplotlib.pyplot as plt  # Import matplotlib library for creating visualizations

# Load the dataset from the specified file path
file_path = '/content/drive/MyDrive/Assignment/Dataset salary 2024.csv'  # Change this to the correct path
data = pd.read_csv(file_path)  # Read the CSV file into a DataFrame

# Display the first few rows of the dataset to understand its structure
print(data.head())

# Create a copy of the original DataFrame to work with
encoded_data = data.copy()

# Map experience levels to numeric values for easier analysis
experience_mapping = {'EN': 1, 'MI': 2, 'SE': 3, 'EX': 4}  # Define a mapping for experience levels
encoded_data['experience_level'] = encoded_data['experience_level'].map(experience_mapping)  # Apply the mapping

# Map employment types to numeric values
employment_mapping = {'FT': 1, 'PT': 2, 'CT': 3, 'FL': 4}  # Define a mapping for employment types
encoded_data['employment_type'] = encoded_data['employment_type'].map(employment_mapping)  # Apply the mapping

# Convert the remote ratio from percentage to a decimal (e.g., 50% becomes 0.5)
encoded_data['remote_ratio'] = encoded_data['remote_ratio'] / 100

# Select only the numerical columns and columns that were encoded
numerical_data = encoded_data[['work_year', 'experience_level', 'employment_type', 'salary_in_usd', 'remote_ratio']]

# Calculate the correlation matrix of the selected numerical data
correlation_matrix = numerical_data.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 6))  # Set the size of the figure
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)  # Create the heatmap with annotations
plt.title('Heatmap of Correlation Between Numerical and Encoded Categorical Columns')  # Set the title of the heatmap
plt.show()  # Display the heatmap
```
