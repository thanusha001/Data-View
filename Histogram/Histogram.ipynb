```python
import pandas as pd  # Import the pandas library for data manipulation
import matplotlib.pyplot as plt  # Import matplotlib for plotting
import seaborn as sns  # Import seaborn for enhanced visualizations

# Load the dataset
file_path = '/content/drive/MyDrive/Assignment/Dataset salary 2024.csv'  # Specify the path to the CSV file
data = pd.read_csv(file_path)  # Read the CSV file into a DataFrame

# Display the first few rows of the dataset
print(data.head())  # Print the first five rows of the DataFrame to check the data

# Histogram of salaries
plt.figure(figsize=(10, 6))  # Create a figure with a specified size
sns.histplot(data['salary_in_usd'], bins=20, kde=True)  # Plot a histogram of the 'salary_in_usd' column with 20 bins and a kernel density estimate
plt.title('Histogram of Salaries')  # Add a title to the histogram
plt.xlabel('Salary in USD')  # Label the x-axis as 'Salary in USD'
plt.ylabel('Frequency')  # Label the y-axis as 'Frequency'
plt.show()  # Display the histogram
```
