```python
import pandas as pd  # Import pandas library for data manipulation
import plotly.express as px  # Import plotly.express for creating interactive visualizations

# Load the dataset from the specified file path
file_path = '/content/drive/MyDrive/Assignment/Dataset salary 2024.csv'  # Path to the CSV file
data = pd.read_csv(file_path)  # Read the CSV file into a DataFrame

# Group the data by 'job_title' and calculate the total salary for each job title
area_data = data.groupby('job_title')['salary_in_usd'].sum().reset_index()  # Aggregate salaries by job title and reset index

# Create an area chart
# Plot 'job_title' on the x-axis and 'salary_in_usd' on the y-axis
# The chart will show the total salaries for each job title
area_chart = px.area(area_data, x='job_title', y='salary_in_usd', title='Area Chart: Total Salaries Across Job Titles')

# Display the area chart
area_chart.show()
```
