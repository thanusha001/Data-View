```python
!pip install pandas plotly  # Install the pandas and plotly libraries if they are not already installed

import pandas as pd  # Import pandas library for data manipulation
import plotly.express as px  # Import plotly.express for creating interactive visualizations

# Load the dataset from the specified file path
file_path = '/content/drive/MyDrive/Assignment/Dataset salary 2024.csv'  # Path to the CSV file
data = pd.read_csv(file_path)  # Read the CSV file into a DataFrame

# Simulate a Gantt chart by adding a 'duration' column
data['duration'] = 1  # Assume each task lasts for 1 year

# Create a 'Start' column by converting the 'work_year' column to a datetime format (assuming start of the year)
data['Start'] = pd.to_datetime(data['work_year'].astype(str) + '-01-01')

# Create a 'Finish' column by adding 1 year to the 'Start' date
data['Finish'] = data['Start'] + pd.DateOffset(years=1)

# Prepare data for the Gantt chart
# Select relevant columns for the Gantt chart and rename them for better readability
gantt_data = data[['job_title', 'Start', 'Finish', 'experience_level']]
gantt_data = gantt_data.rename(columns={'job_title': 'Task', 'experience_level': 'Resource'})

# Plot the Gantt chart using plotly
gantt_chart = px.timeline(gantt_data, x_start="Start", x_end="Finish", y="Task", color="Resource", title='Gantt Chart: Project Timeline')
gantt_chart.show()  # Display the Gantt chart
```
