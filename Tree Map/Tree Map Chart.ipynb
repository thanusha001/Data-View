```python
import pandas as pd
import plotly.express as px

# Load the dataset
file_path = '/content/drive/MyDrive/Assignment/Dataset salary 2024.csv'  # Change this to the correct path
data = pd.read_csv(file_path)  # Read the CSV file into a DataFrame

# Tree Chart (Treemap)
# Group the data by 'job_title' and 'experience_level' and sum the 'salary_in_usd' for each group
tree_data = data.groupby(['job_title', 'experience_level'])['salary_in_usd'].sum().reset_index()

# Create a treemap chart using Plotly Express
# Specify the path to define the hierarchy and the values to determine the size of each segment
tree_chart = px.treemap(
    tree_data,  # Data for the treemap
    path=['job_title', 'experience_level'],  # Define the hierarchy of the treemap
    values='salary_in_usd',  # Define the values for the size of each segment
    title='Tree Chart: Distribution of Salaries Across Job Titles and Experience Levels'  # Title of the chart
)

# Display the treemap chart
tree_chart.show()
```
