```python
import pandas as pd  # Import pandas library for data manipulation
import plotly.express as px  # Import plotly.express for creating interactive visualizations

# Load the dataset from the specified file path
file_path = '/content/drive/MyDrive/Assignment/Dataset salary 2024.csv'  # Path to the CSV file
data = pd.read_csv(file_path)  # Read the CSV file into a DataFrame

# Display the first few rows of the dataset to understand its structure
print(data.head())

# Create a funnel chart to show the distribution of salaries across different job titles
# Group data by 'job_title' and sum up 'salary_in_usd' for each job title
funnel_data = data.groupby('job_title')['salary_in_usd'].sum().reset_index()  # Group by job title and calculate total salary
funnel_data = funnel_data.sort_values(by='salary_in_usd', ascending=False)  # Sort job titles by total salary in descending order

# Plot the funnel chart using plotly
funnel_chart = px.funnel(funnel_data, x='salary_in_usd', y='job_title', title='Funnel Chart: Distribution of Salaries Across Job Titles')

# Display the funnel chart
funnel_chart.show()
```
